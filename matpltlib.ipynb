# matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,20,200)
```

```python
y = np.sin(x)
```

```python
plt.scatter(x,y)
plt.axhline(y=0.75, ls='--', c='r')
plt.text(3.5,0,'sin(x)',weight = 'bold', c='r')
```

```python
import matplotlib.pyplot as plt
import numpy as np
rng = np.random.RandomState(10)
```

```python
k = rng.rand(100)
f = rng.rand(100)
```

```python

plt.scatter(k, f, c=rng.rand(100),sizes=1000*rng.rand(100),alpha=0.2 ,cmap = 'viridis')
```

```python
x = np.random.normal(0,0.8,1000)
y = np.random.normal(-3,1,1000)
z = np.random.normal(3,2.,1000)
```

```python
kwargs = dict(alpha=0.3, bins=40,density=True)
```

```python
plt.hist(x, **kwargs)
plt.hist(y,**kwargs)
plt.hist(z,**kwargs)
```

```python
x = np.random.rand(1000)
```

```python
y = np.random.rand(1000)
```

```python
plt.hist2d(x,y,bins=[10,20])
plt.xlabel('x-ax')
plt.ylabel('y-ax')
plt.colorbar()
```

```python
x = [1,2,3,4,5,6]
y = [2,3,4,5,6,7]
plt.bar(x,y)
plt.title('barbar')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = ['dog','cat','elephant','mouse','human','snake']
plt.pie(x,labels=y, autopct='%1.1f%%')
plt.title('pie')
plt.xlabel('x-label')
plt.ylabel('y-label')
```

```python
x = [1,2,3,4,4,5]
y = [1,2,3,4,5,6]
plt.subplot(2,2,1)
plt.bar(x,y)
plt.title('bar')

x = [1,2,3,4,4,5]
y = [1,2,3,4,5,6]
plt.subplot(2,2,2)
plt.plot(x,y,color = 'r')
plt.title('line')

x = [1,2,3,4,4,5]
y = [1,2,3,4,5,6]
plt.subplot(2,2,3)
plt.pie(x,labels=y,autopct='%1.1f%%')
plt.title('pie')

x = [1,2,3,4,4,5,6,7,8,9]
y = [1,2,3,4,5,6,3,2,5,2]
plt.subplot(2,2,4)
plt.scatter(x,y,color = 'black')
plt.title('scatter')
```
